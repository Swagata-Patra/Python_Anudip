Lab2: Write a python program using Interpolation to fill in missing values in thedata frame. Then generate a subject wise highest score in the form of a bar chart.Input:data =data = {'Subject':data = {'Subject': ['Math', 'Physics', 'Chemistry', 'Biology'],'Student_A': [80, 85, np.nan, 70],'Student_B': [90, np.nan, 75, 85],'Student_C': [80, 88, np.nan, 78]}

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Interpolating missing values
data = {'Subject': ['Math', 'Physics', 'Chemistry', 'Biology'],
        'Student_A': [80, 85, np.nan, 70],
        'Student_B': [90, np.nan, 75, 85],
        'Student_C': [80, 88, np.nan, 78]}
df=pd.DataFrame(data)
print("Oiginal DataFrame:")
print(df)
print("\nDataFrame after Interpolation:")
df.interpolate(inplace=True)
print(df)
```

```python
# Getting subject-wise heighest scores
df['Highest_Score']=df[['Student_A','Student_B','Student_C']].max(axis=1)

# Plotting bar chart for subject-wise heighest scores
plt.figure(figsize=(8,4))
plt.bar(df['Subject'], df['Highest_Score'], color='skyblue')
plt.title('Subject-wise Highest Score')
plt.xlabel('Subject')
plt.ylabel('Highest Score')
plt.show()
```

Out of all the subjects math has the heighest score and chemistry has the lowest score.
